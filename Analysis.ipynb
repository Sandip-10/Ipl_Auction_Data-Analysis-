```python
# Import all requied python libaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
# create and Read the csv file  into a Pandas DataFrame
ipl = pd.read_csv('E:\WorksSpace\Anudip_Assignment\Project_Anudip\ipl_2023_dataset.csv')
```

```python
# Explore the Dataset using .head()
ipl.head()
```

```python
# Find the total number of rows and the total columns in a given dataset using .shape  attribute
ipl.shape
```

```python
# Show the columns name using .columns method
ipl.columns
```

```python
# Check the columns name datatypes and the total non-null objects using .info()
ipl.info()
```

```python
# Drop any unnecessary column using .drop () method
ipl.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Check the column after using drop the 'Unnamed: 0' column remove.
ipl.head()
```

```python
# Check total null value that present in the dataset
ipl.isnull().sum()
```

```python
# Check the 'Cost in $' row that have any NULL or NaN value.
ipl[ipl['Cost in $ (K)'].isnull()]
```

```python
# Fill the NULL or NaN values with zero in 'Cost in Rs. (CR)' and 'Cost in $ (K)' in a given DataFrame
ipl['Cost in Rs. (CR)'] = ipl['Cost in Rs. (CR)'].fillna(0)
ipl['Cost in $ (K)'] = ipl['Cost in $ (K)'].fillna(0)
```

```python
# Check the 2022 squad that have null values.
ipl[ipl['2022 Squad'].isnull()]
```

```python
# Replace the null value with NOt participated text using .fillna () method in pandas
ipl['2022 Squad'] = ipl['2022 Squad'].fillna('Not Participated')
```

```python
# Verify after replace null value with 'NOt participated' text.
ipl.isnull().sum()
```

```python
# Loking all the rows where teams spending 15 cores and those should be unique
teams = ipl[ipl['Cost in Rs. (CR)']>15]['2023 Squad'].unique()
teams
```

```python
# crete a column 'status' with witten sold or unsold players
ipl['Status'] = ipl['2023 Squad'].replace(teams, 'Sold')
ipl.head()
```

```python
# Check Dataset to unsold players
ipl
```

```python
# Check Max,Mean,Min,STD of Cost in Rs. and Cost in $
print(ipl.describe())
```

```python
# Calculate the total money spend on 'Cost in Rs. (CR)' in 2023 Squad.
total_cost = ipl['Cost in Rs. (CR)'].sum()
print("Total cost price of 'Cost in Rs. (CR)':", total_cost)
```

```python
# Find the player name who got maximum price, and 2023 squad name in Rs. (CR) and in $ (K)
max_cost_player_row = ipl.loc[ipl['Cost in Rs. (CR)'].idxmax()]

# Extract player name, cost price, and 2023 squad name
max_cost_player_name = max_cost_player_row['Player Name']
max_cost_price = max_cost_player_row['Cost in Rs. (CR)']
max_cost_price2 = max_cost_player_row['Cost in $ (K)']
squad_2023_name = max_cost_player_row['2023 Squad']

print("Maximum cost price of a player in Rs. (CR) with player name and 2023 Squad name:")
print("Player Name:", max_cost_player_name)
print("Cost in Rs. (CR):", max_cost_price)
print("Cost in $ (K)", max_cost_price2)
print("2023 Squad Name:", squad_2023_name)
```

```python
# Check total number of players by 'base price'
players_count_by_base_price = ipl['Base Price'].value_counts().sort_index()

print("Total Players Count Based on Base Price:")
print(players_count_by_base_price)
```

```python

# Plotting bar plot for total players count based on base price
plt.figure(figsize=(12, 6))
players_count_by_base_price.plot(kind='bar', color='skyblue')
plt.title('Total Players Count Based on Base Price (IPL 2023)')
plt.xlabel('Base Price (CR)')
plt.ylabel('Total Player Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# How many types of players participated in the auction.
types = ipl['Type'].value_counts()
types.reset_index()
```

```python
# Count the number of players for each type
types = ipl['Type'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 6))
types.plot(kind='bar', color='skyblue')
plt.title('Types of Players in the Auction')
plt.xlabel('Player Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Pie Polt of Total cost in Rs. for Batsman, Bowler, All-rounder and Wicketkeeper 
total_cost_by_type = ipl.groupby('Type')['Cost in Rs. (CR)'].sum()

# Create pie plot
plt.figure(figsize=(8, 8))
plt.pie(total_cost_by_type, labels=total_cost_by_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Cost in Rs. (CR) by Type')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Check sold and unsold player of the daaset
ipl.groupby('Status')['Player Name'].count()
```

```python
# Create a bar graph for sold and unsold players
plt.figure(figsize=(20,5))
fig = sns.countplot(x='Status', data=ipl, hue='Status')
plt.xlabel('Sold or Unsold')
plt.ylabel('Number of Players')
plt.title('Sold vs Unsold Players', fontsize=15)
plt.show()


```

```python
# Count total number of players bought by each team
players_bought_by_team = ipl[ipl['2023 Squad'] != 'Unsold']['2023 Squad'].value_counts()
print(players_bought_by_team)
```

```python
# Total number of players bought by each team.

plt.figure(figsize=(20,10))
fig = sns.countplot(x='2023 Squad', data=ipl[ipl['2023 Squad'] != 'Unsold'])
plt.xlabel('Number of Players')  
plt.ylabel('Teams Name')  
plt.title('Players Bought by Each Team', fontsize=12)
plt.xticks(rotation=70) 
plt.show()


```

```python
# Check total retained players

# Filter out players who are retained.
retained_players = ipl[ipl['Cost in Rs. (CR)'] == 0]

# Count the total number of retained players
total_retained_players = retained_players.shape[0]

# Count the number of retained players per team
retained_players_per_team = retained_players.groupby('2023 Squad').size()


print("Total retained players count:", total_retained_players)
print("\nRetained players per team:")
print(retained_players_per_team)

```

```python
# Create a bar plot for retained players per team
plt.figure(figsize=(12, 6))
retained_players_per_team.plot(kind='bar', color='skyblue')
plt.title('Retained Players per Team (IPL 2023)')
plt.xlabel('Team')
plt.ylabel('Number of Retained Players')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting the count of retained players per team as a pie plot
plt.figure(figsize=(8, 8))
plt.pie(retained_players_per_team, labels=retained_players_per_team.index, autopct='%1.1f%%', startangle=140)
plt.title('Retained Players Distribution per Team (IPL 2023)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Check number of players based on their types and check which team has how many of them.

# Group the data by '2023 Squad' and 'Type', and then count the number of players
players_count_by_type_and_team = ipl.groupby(['2023 Squad', 'Type']).size().unstack(fill_value=0)

# Print the count of players based on their types and team
print("Number of players based on their types and team:")
print(players_count_by_type_and_team)
```

```python
# Plotting separate bar plots for each player type
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# All-rounder
players_count_by_type_and_team['ALL-ROUNDER'].plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Count of All-rounders')
axes[0, 0].set_xlabel('Team')
axes[0, 0].set_ylabel('Number of Players')

# Batsman
players_count_by_type_and_team['BATSMAN'].plot(kind='bar', ax=axes[0, 1], color='orange')
axes[0, 1].set_title('Count of Batsmen')
axes[0, 1].set_xlabel('Team')
axes[0, 1].set_ylabel('Number of Players')

# Bowler
players_count_by_type_and_team['BOWLER'].plot(kind='bar', ax=axes[1, 0], color='green')
axes[1, 0].set_title('Count of Bowlers')
axes[1, 0].set_xlabel('Team')
axes[1, 0].set_ylabel('Number of Players')

# Wicketkeeper
players_count_by_type_and_team['WICKETKEEPER'].plot(kind='bar', ax=axes[1, 1], color='red')
axes[1, 1].set_title('Count of Wicketkeepers')
axes[1, 1].set_xlabel('Team')
axes[1, 1].set_ylabel('Number of Players')

plt.tight_layout()
plt.show()
```
